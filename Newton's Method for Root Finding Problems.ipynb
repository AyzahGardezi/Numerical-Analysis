Newton's Method for Root Finding Problems

```python
import numpy as np
from sympy import symbols, exp, diff, lambdify
import matplotlib.pyplot as plt
```

```python
x = symbols('x')

f = exp(x) - 2

f_prime = diff(f, x, 1)

f_numeric = lambdify(x, f, 'numpy')
f_prime_numeric = lambdify(x, f_prime, 'numpy')

x0 = 0.5 
guesses = [x0]
```

```python
for i in range(0, 100):
    x1 = x0 - (f_numeric(x0) / f_prime_numeric(x0))
    guesses.append(x1)

    if (np.abs(x1 - x0) < 10**(-7)):
        break
        
    x0 = x1

print(guesses)
```

```python

x_vals = np.linspace(-1, 2, 400)
y_vals = f_numeric(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='$f(x) = e^x - 2$', color='blue')
plt.scatter(guesses, f_numeric(np.array(guesses)), color='red', label='Iterations', zorder=5)

plt.title("Newton's Method for $f(x) = e^x - 2$")
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.legend()
plt.grid()
plt.ylim(-3, 3)
plt.xlim(-1, 2)

plt.show()
```
