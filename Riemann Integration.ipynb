Integral of 2x^2 + 3x + 3 with limits [-1, 1]:

```python
correct_value = 22/3
print(correct_value)
```

Defining variables and function:

```python
a = -1
b = 1
n = 1000

def f(x):
  return 2 * x**2 + 3*x + 3
```

### Riemann Integration

```python
def riemann_integration(a, b, f, n):
  res = 0
  for i in range(1, n - 1):
    res = res + f(a + (i * (b - a) / n))
  res = ((b - a) / n) * res
  return res

approximated_value = riemann_integration(a, b, f, n)
print(approximated_value)
```

```python
error = correct_value - approximated_value
print(error)
```

# Error Analysis

```python
n_iterations = list(range(1, 101))
```

```python
approx_values = []

for m in n_iterations:
  approx_values.append(riemann_integration(a, b, f, m))

print(approx_values)
```

```python
import numpy as np

approx_values_array = np.array(approx_values)

error_values = np.abs((correct_value - approx_values_array)/correct_value)
```

```python
import matplotlib.pyplot as plt

plt.title('Riemann Integration Error %')
plt.xlabel('number of iterations')
plt.ylabel('error')

plt.plot(n_iterations, error_values)
```
