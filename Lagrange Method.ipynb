Task: Lagrange Method

```python
import numpy as np
from sympy import symbols
import matplotlib.pyplot as plt

# x = symbols('x')
X = np.array([0, 2, 4, 6, 8])
y = np.array([0, 5, 4, 2, 3])
```

```python
def lagrange(x, X, y):
    total = 0.0
    n = len(X)
    
    for k in range(n):
        li = 1.0
        for i in range(n):
            if k != i:
                li *= (x - X[i]) / (X[k] - X[i])
        total += li * y[k]
    
    return total
    
x_values = np.linspace(-1, 9, 500)
y_values = [lagrange(x, X, y) for x in x_values]

plt.plot(X, y, 'ro', label='Data Points')
plt.plot(x_values, y_values, 'b-', label='Lagrange Interpolation')
plt.title('Lagrange Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-1, 9)
plt.legend()
plt.grid()
plt.show()

```
