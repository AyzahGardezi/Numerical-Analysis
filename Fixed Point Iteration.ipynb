Task: Implement Fixed Point Iteration

```python
import numpy as np
from sympy import diff, symbols, cos, factorial, lambdify
import matplotlib.pyplot as plt
```

```python
def g(x):
    return np.cos(x)
```

```python
x0 = 0
max_iterations = 100
tolerance = 1e-7
```

```python
approximations = [x0]
```

```python

for _ in range(max_iterations):
    x1 = g(approximations[-1])
    approximations.append(x1)
    
    if abs(x1 - approximations[-2]) < tolerance:
        break

root = approximations[-1]
print(f"Found root: {root}")
```

```python
x_values = np.linspace(-1, 2, 400)
g_values = g(x_values)
f_values = g_values - x_values  # f(x) = cos(x) - x

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(x_values, g_values, label='g(x) = cos(x)', color='blue')
plt.scatter(approximations, g(np.array(approximations)), color='red', label='Iterations')
plt.title('Fixed Point Iteration for g(x)')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x_values, f_values, label='f(x) = cos(x) - x', color='green')
plt.plot(x_values, g_values, label='g(x) = cos(x)', color='blue')
plt.scatter(approximations, g(np.array(approximations)), color='red', label='Iterations')
plt.title('Function f(x) and g(x)')
plt.xlabel('x')
plt.ylabel('f(x) and g(x)')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.legend()

plt.show()
```
